# Алгоритм GrubCut
http://en.wikipedia.org/wiki/GrabCut

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Загружаем изображение

```python
img = cv2.imread('../data/messi5.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
mask = np.zeros(img.shape[:2], np.uint8)  # выделяем память под результат работы алгоритма

# выделяем память необходимую для работы алгоритма
bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)

rect = (50,50,450,290)  # прямоугльником задаем область принадлежащую объекту
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask==2)|(mask==0), 0, 1).astype('uint8')
result = img * mask2[:,:,np.newaxis]
```

# Выводим результат

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```
